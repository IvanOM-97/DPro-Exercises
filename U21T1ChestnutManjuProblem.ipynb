```python
'''
EL PROBLEMA DE BOLLO DE CASTAÑA
  Una herramienta llamada "Baybayin" que aparece en el manga "Doraemon"
  tiene un efecto misterioso que un determinado objeto duplica cada 5 minutos.
  Como episodio relacionado con esta herramienta,
  muestra cómo el número de bollos de castañas aumenta constantemente.
  En la película, no podía comerme todos los bollos de castañas que habían empezado a aumentar,
  así que tuve que tirarlos al espacio.

  A partir de este episodio, esta vez vamos a calcular
  cómo sigue aumentando el volumen de un panecillo de castañas.
'''
```

```python
'''
PROBLEMA 1: EL DIA EN QUE LOS BOLLOS DE CASTAÑA CUBREN EL SISTEMA SOLAR
  Si espolvorea un solo bollo de castaña con "Baybayin",
  ¿cuánto tiempo estará cubierto el sistema solar con bollos de castaña?
  Crea un programa para calcular esto.

  Calcule después de asumir el volumen de bollos de castaña y el sistema solar.
  Puede simplificar el problema al no considerar la interaccion entre los bollos de castaño y la gravedad.
  Cree un programa de propósito general que pueda aplicarse a otras situaciones,
  como cuando el Tokyo Dome se llena cuando rocía un Byvine en la pelota.
  Además, haga que la computadora portátil sea fácil de entender, como mostrar gráficos.
'''

import matplotlib.pyplot as plt

def calculate_duplication_time(initial_volume, max_volume, interval_minutes):
    """
    Simulate the growth of volume by duplication every fixed interval.
      Parameters:
          initial_volume (float): Starting volume in cubic meters
          max_volume (float): Target volume to reach or exceed
          interval_minutes (int): Minutes between each duplication

      Returns:
          total_minutes (int): Total time in minutes to reach/exceed max_volume
          volumes (list): Volume progression at each interval
          times (list): Time progression in minutes
    """
    current_volume = initial_volume
    time = 0
    volumes = [current_volume]
    times = [time]

    while current_volume < max_volume:
        current_volume *= 2
        time += interval_minutes
        volumes.append(current_volume)
        times.append(time)

    return time, volumes, times

# PARAMETERS
init_volume = 1e-4                       # Volume of one chestnut bun in cubic meters
duplication_interval_min = 5                # Duplication every 5 minutes

# Volume of the solar system (approximate sphere up to Pluto's orbit ~40 AU)
# Solar_system_radius = 1 AU = 1.496e11 m → 40 AU ≈ 6e12 m
# Solar_system_volume = (4/3) * 3.141592653589793 * (Solar_system_radius ** 3) ≈ 9e38 m^3

objetive_volume = 9e38                      # Solar system volume in cubic meters

# Run the simulation
time_required, volume_growth, time_points = calculate_duplication_time(
    init_volume, objetive_volume, duplication_interval_min)

# Convert time from minutes to more understandable units
total_days = time_required / (60 * 24)
total_years = total_days / 365.25

# Display results
print("CHESTNUT BUN EXPLOSION SIMULATION")
print(f"Total Time: {time_required} minutes")
print(f"≈ {total_days:.2f} days")
print(f"≈ {total_years:.2f} years")
print(f"Final Volume: {volume_growth[-1]:.2e} m³ (≥ Solar System Volume)")

# Plot the growth
plt.figure(figsize=(12, 6))
plt.plot(time_points, volume_growth, label="Chestnut Bun Volume", color="brown")
plt.axhline(objetive_volume, color="red", linestyle="--", label="Solar System Volume")
plt.yscale("log")
plt.xlabel("Time (minutes)")
plt.ylabel("Volume (m^3)")
plt.title("Growth of Chestnut Buns over Time")
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.tight_layout()
plt.show()
```
