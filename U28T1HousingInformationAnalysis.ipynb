```python
'''
PROBLEMA 5 - INVESTIGACION TERMINOLOGICA
  Es importante investigar y comprender los siguientes términos.
  Esto permitirá captar la naturaleza y las características de los datos.
    - Curtosis: Indicador que representa el grado de concentración de una distribución.
    - Asimetría: Indicador que muestra la falta de simetría en una distribución.
'''
import pandas as pd
import missingno as msno
from scipy.stats import kurtosis, skew

# Load the data
data = pd.read_csv('train.csv')

# Drop columns that have 5 or more missing values
data_cleaned = data.dropna(axis=1, thresh=len(data) - 4)

# Drop rows that still contain any missing values
data_cleaned = data_cleaned.dropna()

# Select only the numeric columns
# We calculate kurtosis and skewness only for numeric data
numeric_data = data_cleaned.select_dtypes(include='number')

# Calculate kurtosis for each numeric column
# Kurtosis shows how peaked the distribution is
print("Kurtosis of each numeric feature:")
print(numeric_data.kurtosis())

# Calculate skewness for each numeric column
# Skewness shows how asymmetric the distribution is
print("\nSkewness of each numeric feature:")
print(numeric_data.skew())
```

```python
'''
PREGUNTA 6 - COMPROBACION DE LA DISTRIBUCION
    - Usemos sns.displot() o sns.histplot() de la librería seaborn para visualizar la distribución de la variable objetivo, y calculemos también la curtosis y la asimetría.
    - Realizaremos una transformación logarítmica sobre la variable objetivo.
    - Luego, visualizaremos la distribución después de la transformación logarítmica, y también calcularemos la curtosis y la asimetría tras dicha transformación.

  Explicaremos cómo cambia la distribución antes y después de la transformación logarítmica, utilizando el término distribución normal.
  Este tipo de transformación puede, en algunos casos, contribuir a mejorar la precisión predictiva del modelo.
'''
import pandas as pd
import seaborn as sns
import missingno as msno
import matplotlib.pyplot as plt
from scipy.stats import kurtosis, skew
import numpy as np

# Load the data
data = pd.read_csv('train.csv')

# Drop columns that have 5 or more missing values
data_cleaned = data.dropna(axis=1, thresh=len(data) - 4)

# Drop rows that still contain any missing values
data_cleaned = data_cleaned.dropna()

# Select only the numeric columns
# We calculate kurtosis and skewness only for numeric data
numeric_data = data_cleaned.select_dtypes(include='number')

# Check the distribution of the manifest variables
sns.histplot(data['SalePrice'], kde=True)

# Calculate kurtosis and skewness
print(f"Kurtosis: {data['SalePrice'].kurtosis()}, Skewness: {data['SalePrice'].skew()}")

# Logarithmic transformation
log_price = np.log(data['SalePrice'])

# Check the distribution after logarithmic transformation
sns.histplot(log_price, kde=True)

# Calculate kurtosis and skewness
print(f"Log Kurtosis: {log_price.kurtosis()}, Log Skewness: {log_price.skew()}")
```

```python
'''
PREGUNTA 7 - COMPROBACION DEL COEFICIENTE DE CORRELACION
  Verificar la correlación entre cada característica para identificar qué variables están fuertemente relacionadas con la variable objetivo.
    - Crear un mapa de calor de la matriz de correlación para investigar relaciones con alta correlación.
    - Identificar las 10 características con mayor correlación con la variable objetivo y mostrar su matriz de correlación mediante un mapa de calor.
    - Explicar en qué representa cada una de las características seleccionadas, basándose en la descripción de datos de Kaggle.
    - Encontrar 3 pares de variables entre las seleccionadas que tengan una alta correlación entre sí.

  Mediante esta verificación, se puede identificar variables con fuerte correlación mutua, y así evitar que afecten en exceso los resultados del análisis.
'''
import pandas as pd
import seaborn as sns
import missingno as msno
import matplotlib.pyplot as plt
import numpy as np

from scipy.stats import kurtosis, skew

# Load the data
data = pd.read_csv('train.csv')

# Drop columns that have 5 or more missing values
# and drop rows that still contain any missing values
data_cleaned = data.dropna(axis=1, thresh=len(data) - 4)
data_cleaned = data_cleaned.dropna()

# Select only the numeric columns
numeric_data = data_cleaned.select_dtypes(include='number')

# Calculate the correlation matrix
corr = numeric_data.corr()

# Display the heatmap
plt.figure(figsize=(15, 15))
sns.heatmap(corr, cmap='coolwarm', square=False, annot=False, fmt=".2f")
plt.show()

# Display the top 10 highly correlated features
top10_corr = corr.nlargest(10, 'SalePrice')['SalePrice'].index
plt.figure(figsize=(10, 10))
sns.heatmap(data[top10_corr].corr(), annot=True, cmap='coolwarm', square=True)
plt.show()
```
