```python
'''
PROBLEMA 1 - COMPARTE Y EJECUTA EL MODELO TUTORIAL OFICIAL
  https://www.tensorflow.org/tutorials?hl=es-419

'''
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
import numpy as np

# Cargado de datos
fashion_mnist = tf.keras.datasets.fashion_mnist
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

# Normalizacion
x_train, x_test = x_train / 255.0, x_test / 255.0

# Construyendo el modelo
model = models.Sequential([layers.Flatten(input_shape=(28, 28)),
                           layers.Dense(128, activation='relu'),
                           layers.Dense(10)
                           ])

# Compilacion y entrenamiento
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=5, validation_split=0.1)

# Evaluacion
model.evaluate(x_test,  y_test)
```

```python
'''
PROBLEMA 3 - APRENDIENDO IRIS (CLASIFICACION BINARIA) CON KERAS
  Rewrite the TensorFlow binary classification for the Iris dataset in Keras.
'''
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.datasets import load_iris

# Cargando el dataset
df = pd.read_csv('Iris.csv')

# Preprocesamiento
df = df[df['Species'].isin(['Iris-vericolor', 'Iris-virginica'])]
X = df[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]].values
y = (df['Species'] == 'Iris-virginica').astype(int).values        # Binary: virginica = 1

# Division
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Construyendo el modelo
model = Sequential([Dense(10, activation='relu', input_shape=(4,)),
                    Dense(1, activation='sigmoid')
                    ])

# Compilacion y entrenamiento
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=50, validation_split=0.1)
model.evaluate(X_test, y_test)

```

```python
'''
PROBLEMA 4 - APRENDIENDO IRIS (CLASIFICACION MULTICLASE) CON KERAS
  Rewrite the TensorFlow multi-class classification for the Iris dataset in Keras.
'''
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical

# Cargando el dataset considerando las 3 clases
df = pd.read_csv('Iris.csv')
X = df[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]].values
y = df['Species'].values

# Preprocesamiento
le = LabelEncoder()
y = le.fit_transform(y)
y = to_categorical(y)

# Division
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Construyendo el modelo
model = Sequential([Dense(16, activation='relu', input_shape=(4,)),
                    Dense(3, activation='softmax')
                    ])

# Compilacion y entrenamiento
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=50, validation_split=0.1)
model.evaluate(X_test, y_test)
```

```python
'''
PROBLEMA 5 - APRENDIENDO HOUSE PRICES CON KERAS
  Translate the TensorFlow regression on the House Prices dataset into Keras.
'''
df = pd.read_csv("train.csv")
df = df[["GrLivArea", "YearBuilt", "SalePrice"]].dropna()

X = df[["GrLivArea", "YearBuilt"]].values
y = df["SalePrice"].values

# Normalizando entradas
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Division train-test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Construccion del modelo
model = Sequential([Dense(64, activation='relu', input_shape=(2,)),
                    Dense(1)
                    ])

# Compilacion y entrenamiento
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
model.fit(X_train, y_train, epochs=100, validation_split=0.1)
model.evaluate(X_test, y_test)
```

```python
'''
PROBLEMA 6 - APRENDIENDO MNIST CON KERAS
  Please rewrite the TensorFlow multi-class image classification using the MNIST dataset in Keras.
'''
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# Cargando el dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Preprocesamiento
x_train = x_train.reshape(-1,784)/ 255.0
x_test = x_test.reshape(-1,784)/ 255.0
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

# Construyendo el modelo
model = Sequential([Dense(128, activation='relu', input_shape=(784,)),
                    Dense(10, activation='softmax')
                    ])

# Compilacion y entrenamiento
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=10, validation_split=0.1)
model.evaluate(x_test, y_test)
```

```python
'''
PROBLEMA 7 - REESCRIBIENDO A PyTorch (TAREA AVANZADA)
  Rewrite the four types of problems in PyTorch.
  https://docs.pytorch.org/tutorials/index.html
'''
# 7.1 - CLASIFICACION BINARIA EN IRIS
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# cargado y preprocesamiento
df = pd.read_csv('Iris.csv')
df = df[df['Species'].isin(['Iris-vericolor', 'Iris-virginica'])]
X = df[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]].values
y = (df['Species'] == 'Iris-virginica').astype(int).values

scaler = StandardScaler()
X = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train, y_train = torch.tensor(X_train, dtype=torch.float32), torch.tensor(y_train, dtype=torch.float32).unsqueeze(1)
X_test, y_test = torch.tensor(X_test, dtype=torch.float32), torch.tensor(y_test, dtype=torch.float32).unsqueeze(1)

# Modelo
model = nn.Sequential(nn.Linear(4, 10),
                      nn.ReLU(),
                      nn.Linear(10, 1),
                      nn.Sigmoid()
                      )
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

# Training loop
for epoch in range(50):
    model.train()
    y_pred = model(X_train)
    loss = criterion(y_pred, y_train)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

# Evaluacion
with torch.no_grad():
    model.eval()
    test_pred = model(X_test)
    acc = ((test_pred > 0.5) == y_test).float().mean()
    print(f"Test Accuracy for IRIS binary classification: {acc.item():.4f}")
    print("\n")

# 7.2 CLASIFICACION MULTICLASE EN IRIS
from sklearn.preprocessing import LabelEncoder
from torch.utils.data import TensorDataset, DataLoader

# Cargando dataset
df = pd.read_csv('Iris.csv')
X = df[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]].values
y = df['Species'].values

le = LabelEncoder()
y = le.fit_transform(y)

scaler = StandardScaler()
X = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

X_train, y_train = torch.tensor(X_train, dtype=torch.float32), torch.tensor(y_train, dtype=torch.long)
X_test, y_test = torch.tensor(X_test, dtype=torch.float32), torch.tensor(y_test, dtype=torch.long)

# Modelo
model = nn.Sequential(nn.Linear(4, 16),
                      nn.ReLU(),
                      nn.Linear(16, 3)
                      )

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

# Entrenamiento
for epoch in range(50):
    model.train()
    optimizer.zero_grad()
    y_pred = model(X_train)
    loss = criterion(y_pred, y_train)
    loss.backward()
    optimizer.step()

# Evaluación
with torch.no_grad():
    model.eval()
    preds = torch.argmax(model(X_test), dim=1)
    acc = (preds == y_test).float().mean()
    print(f"Test Accuracy for IRIS multi-class classification: {acc.item():.4f}")
    print("\n")

# 7.3 REGRESION HOUSE PRICES
df = pd.read_csv("train.csv")
df = df[["GrLivArea", "YearBuilt", "SalePrice"]].dropna()

X = df[["GrLivArea", "YearBuilt"]].values
y = df["SalePrice"].values.reshape(-1, 1)

scaler = StandardScaler()
X = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train, y_train = torch.tensor(X_train, dtype=torch.float32), torch.tensor(y_train, dtype=torch.float32)
X_test, y_test = torch.tensor(X_test, dtype=torch.float32), torch.tensor(y_test, dtype=torch.float32)

# Modelo
model = nn.Sequential(nn.Linear(2, 64),
                      nn.ReLU(),
                      nn.Linear(64, 1)
                      )

criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

# Entrenamiento
for epoch in range(100):
    model.train()
    optimizer.zero_grad()
    y_pred = model(X_train)
    loss = criterion(y_pred, y_train)
    loss.backward()
    optimizer.step()

# Evaluación
with torch.no_grad():
    model.eval()
    y_pred = model(X_test)
    mae = torch.mean(torch.abs(y_pred - y_test))
    print(f"Mean Absolute Error (House Prices): {mae.item():.2f}")
    print("\n")

# 7.4 MULTI-CLASS IMAGE CLASSIFICATION
from tensorflow.keras.datasets import mnist
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# Transformaciones y dataset
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])

train_dataset = datasets.FashionMNIST(root='./data', train=True, download=True, transform=transform)
test_dataset = datasets.FashionMNIST(root='./data', train=False, download=True, transform=transform)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)

# Modelo
model = nn.Sequential(nn.Flatten(),
                      nn.Linear(28*28, 128),
                      nn.ReLU(),
                      nn.Linear(128, 10)
                      )

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Entrenamiento
for epoch in range(5):
    model.train()
    running_loss = 0.0
    for images, labels in train_loader:
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
    print(f"Epoch {epoch+1}, Loss: {running_loss/len(train_loader):.4f}")

# Evaluación
with torch.no_grad():
    model.eval()
    correct, total = 0, 0
    for images, labels in test_loader:
        outputs = model(images)
        _, preds = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (preds == labels).sum().item()
    print(f"Test Accuracy for Fashion MNIST: {100 * correct / total:.2f}%")
```
