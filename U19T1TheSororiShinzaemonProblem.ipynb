```python
'''
PROBLEMA 1: calcular el nro de granos de arroz en el dia 100
'''
import matplotlib.pyplot as plt
import numpy as np

num_grain = 0              #nro de granos de arroz en el dia 1
num_days = 100                #nro de dias
grain_list = []            #lista del nro de granos de arroz
day_list = []
total_grain = num_grain

for i in range(1, num_days + 1):
    num_grain = 2 ** (i-1)
    total_grain += num_grain
    grain_list.append(num_grain)

print(f"el dia {num_days} se recibieron {num_grain} granos de arroz")
print(f"hasta el dia {num_days} se recibieron {total_grain} granos de arroz en total")

#visualizacion de calculo
fig, (ax1, ax2) = plt.subplots(nrows= 1, ncols= 2, figsize=(18, 6))
plt.subplots_adjust(wspace=0.25, hspace=0.25, )

ax1.set_title("Cantidad de granos", color = "red")
ax1.grid(True)
ax1.set_xlabel("nro de dia", )
ax1.set_xticks(np.arange(1, num_days + 1, step = 5))
ax1.set_ylabel("cantidad de arroz [granos]",)
ax1.plot(range(1, num_days + 1), grain_list)

ax2.set_title("Cantidad de granos", color = "red")
ax2.grid(True)
ax2.set_xlabel("nro de dia",)
ax2.set_xticks(np.arange(1, num_days + 1, step = 10))
ax2.set_ylabel("cantidad de arroz [granos]",)
ax2.set_yscale("log", base = 10)
ax2.plot(range(1, num_days + 1), grain_list)

plt.show()
```

```python
'''
PROBLEMA 2: numero de granos de arroz que no sean en el dia 100
            - Crea una funcin con el nmero de das como argumento que pueda utilizarse para cualquier historia
            - responde cuntos granos de arroz recibiras por el nmero de das del curso.
            - Haz una grfica lineal como en la pregunta 1.
            - escriba los comentarios de forma que tambin sean fciles de entender
'''

import matplotlib.pyplot as plt
import numpy as np

def calculate_rice(days):
    """
    Calculates the amount of rice received daily and the cumulative total.
      Parameters:
      days (int): Number of days the history lasts.

      Returns:
      tuple: (list_rice_day, list_rice_total)
      - list_rice_day: List with the number of grains received per day.
      - list_rice_total: List with the cumulative total of grains for each day.
    """

    rice_day = []       # Granos recibidos cada da
    rice_total = []     # Total acumulado hasta ese da
    total = 0           # Inicializacin del acumulado

    # Clculo para cada da
    for day in range(1, days + 1):
        rice_today = 2 ** (day - 1)   # Se duplica cada iniciando con 1 grano
        total += rice_today
        rice_day.append(rice_today)
        rice_total.append(total)

    return rice_day, rice_total

def graph_rice(rice_day, rice_total, days):
    """
    Graphs the amount of rice per day and the running total.
      Parameters:
      rice_day (list): Rice grains received each day.
      rice_total (list): Running total of grains up to each day.
      days (int): Number of days graphed.
    """

    plt.figure(figsize=(12, 6))
    plt.plot(range(1, days + 1), rice_day, label="Arroz por da", color='orange')
    plt.plot(range(1, days + 1), rice_total, label="Total acumulado", color='green')
    plt.xlabel("Da")
    plt.xticks(np.arange(1, days + 1, step = 1))
    plt.ylabel("Cantidad de arroz (granos)")
    plt.title(f"Transicin del arroz recibido y acumulado durante {days} das")
    plt.legend()
    plt.grid(True, which="both", linestyle="--", linewidth=0.5)
    plt.tight_layout()
    plt.show()

# Example of use
days_curso = 10  # se puede cambiar a 81, 100, etc.

# Calculate
rice_day, rice_total = calculate_rice(days_curso)

# Show key results
print(f"Granos recibidos en el da {days_curso}: {rice_day[-1]:,}")
print(f"Total acumulado en {days_curso} das: {rice_total[-1]:,} granos")

# Graphic
graph_rice(rice_day, rice_total, days_curso)
```

```python
'''
PROBLEMA 3: cuantas personas pueden vivir con el arroz que reciben
          - crear un funcion que calcule cuantas personas puedan vivir con el arroz que recibe
              - entradas de la funcion
                  Nmero de granos de arroz (int)
                  Nmero de personas (int)
              - salidas de las funcion
                  Das para vivir (int)
'''

def dias_de_vida(total_grains, number_of_people):
    """
    Calculate how many days a certain number of people can live on a given amount of rice.
      Parameters:
      total_grains (int): Total available rice grains.
      number_people (int): Number of people who will consume the rice.
      Returns:
      int: Number of days all those people can live on the available rice.
      Assumes each person consumes 7,500 grains of rice per day.
    """

    # Each person consumes approximately 7,500 grains of rice per day.
    consumption_per_person = 7500

    # Total daily intake for all people
    total_daily_consumption = number_of_people * consumption_per_person

    # Calculating the number of possible days
    possible_days = total_grains // total_daily_consumption

    return possible_days


# Example of use:
available_grains = 100000  # Total de arroz acumulado en 30 das (2^30 - 1)
people_in_class = 5           # Puedes modificar este nmero

dias = dias_de_vida(available_grains, people_in_class)

print(f"{people_in_class} personas pueden vivir durante {dias} das con {available_grains:,} granos de arroz.")
```
