```python
'''
PROBLEMA 5 - EXTRACCION DE DATOS NECESARIOS
  Para saber cómo extraer de manera eficiente datos específicos de un DataFrame, haga lo siguiente:
    - Extraer columnas de dos maneras diferentes (sepal_width)
    - Extraer datos del 50 al 99
    - Extraer datos de las columnas "petal_length" del 50 al 99
    - Extraer datos cuyo valor es "petal_width = 0.2"
  Para ello, utilice .iloc y .loc. Es importante comprender la diferencia entre ambos.
  La diferencia radica en que se accede por nombre de etiqueta, mientras que se accede por posición.
'''
# Import required libraries
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Store the features (explanatory variables) in a DataFrame called X
X = pd.DataFrame(iris.data, columns=iris.feature_names)

# Store the target variable (species) in a DataFrame called Y
Y = pd.DataFrame(iris.target, columns=["species"])

# Combine the feature data (X) and the target data (Y) into one DataFrame
df = pd.concat([X, Y], axis=1)

# Extract 'sepal_width' column using two different ways
print("Extracting 'sepal_width' using iloc:")
print(df.iloc[:,[1]])

print("\nExtracting 'sepal_width' using loc:")
print(df.loc[:,'sepal width (cm)'])

# Extract rows from index 50 to 99 using iloc (position-based)
print("\nRows from index 50 to 99:")
print(df.iloc[50:99])

# Extract 'petal_length' column from rows 50 to 99 using loc (label-based)
print("\n'petal_length' from rows 50 to 99:")
print(df.loc[50:99, 'petal length (cm)'])

# Extract rows where 'petal_width' is equal to 0.2
print("\nRows where 'petal_width' equals 0.2:")
print(df.loc[df['petal width (cm)'] == 0.2, ['petal width (cm)'] ])
```

```python
'''
PROBLEMA 6 - CREANDO UN DIAGRAMA
  La visualización de datos le ayuda a comprender intuitivamente los resultados de su análisis. .
  Cree el siguiente diagrama:
    - Muestra el número de muestras para cada variable objetivo en un gráfico circular, incluidos los porcentajes
    - Para cada característica, visualice la distribución de datos por variable objetivo utilizando un diagrama de caja.
    - Para cada característica, visualice la distribución de datos para cada variable objetivo utilizando un gráfico de violín.

  matplotlib.pyplot.pie
  matplotlib.pyplot.boxplot
  matplotlib.pyplot.violinplot
'''


import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Store the features (explanatory variables) in a DataFrame called X
X = pd.DataFrame(iris.data, columns=iris.feature_names)

# Store the target variable (species) in a DataFrame called Y
Y = pd.DataFrame(iris.target, columns=["species"])

# Combine the feature data (X) and the target data (Y) into one DataFrame
df = pd.concat([X, Y], axis=1)

# Show the number of samples for each target variable using a pie chart
# Count how many samples there are for each class in the 'target' column
target_counts = df['species'].value_counts()

# Create a pie chart with percentages
plt.figure(figsize=(6, 6))
plt.pie(target_counts, labels=target_counts.index, autopct='%1.1f%%')
plt.title('Sample distribution by target (Pie Chart)')
plt.show()


# Create boxplots for each feature by target variable
# First, make a figure with multiple subplots
features = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
plt.figure(figsize=(12, 8))

for i, feature in enumerate(features):
    plt.subplot(2, 2, i+1)

    # Prepare data grouped by target
    data = [df[df['species'] == label][feature] for label in df['species'].unique()]

    # Make boxplot
    plt.boxplot(data, labels=df['species'].unique())
    plt.title(f'{feature} - Boxplot by Target')
    plt.ylabel(feature)

plt.tight_layout()
plt.show()


# Create violin plots for each feature by target variable
plt.figure(figsize=(12, 8))

for i, feature in enumerate(features):
    plt.subplot(2, 2, i+1)

    # Prepare data grouped by target
    data = [df[df['species'] == label][feature] for label in df['species'].unique()]

    # Make violin plot
    plt.violinplot(data)
    plt.title(f'{feature} - Violin Plot by Target')
    plt.xticks([1, 2, 3], df['species'].unique())
    plt.ylabel(feature)

plt.tight_layout()
plt.show()
```

```python
'''
PROBLEMA 7 - COMPROBACION DE LA RELACION ENTRE CARACTERISTICAS
  Explora las relaciones entre las características. Probando lo siguiente:
    - Crea 1 scatter plot de 6 combinaciones de las 4 características, codificando por color cada especie.
    - Crea 1 matriz de scatter plot que muestre todas las combinaciones de características a la vez.
    - Crea 1 matriz de coeficientes de correlación para las 4 características utilizando métodos DataFrame.
    - Crear un mapa de calor de la matriz de coeficientes de correlación
'''
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Store the features (explanatory variables) in a DataFrame called X
X = pd.DataFrame(iris.data, columns=iris.feature_names)

# Store the target variable (species) in a DataFrame called Y
Y = pd.DataFrame(iris.target, columns=["species"])

# Combine the feature data (X) and the target data (Y) into one DataFrame
df = pd.concat([X, Y], axis=1)

# Scatter plot of two features (petal_length vs petal_width) for all 3 species
# Each species will have a different color
plt.figure(figsize=(6, 4))
sns.scatterplot(data=df, x='sepal length (cm)', y='sepal width (cm)', hue='species')
plt.title('Scatter Plot: Sepal Length vs Sepal Width by Species')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend(title='Species')
plt.show()


# Pairplot (scatter matrix) for all feature combinations, color-coded by species
# This shows how features relate to each other
sns.pairplot(df, hue='species')
plt.suptitle('Pairplot of Iris Dataset Features', y=1.02)
plt.show()


# Correlation matrix using pandas
# This shows how strongly each feature is related to the others (between -1 and 1)
correlation_matrix = df.drop(columns='species').corr()
print(correlation_matrix)


# Heatmap of the correlation matrix
# It visually shows the correlation values using colors
plt.figure(figsize=(6, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
