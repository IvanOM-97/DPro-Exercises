```python
'''
PROBLEMA 1 - CREAR NUMEROS ALEATORIOS
  Cree 500 números aleatorios con una distribución normal bidimensional
  con una media de (-3, 0) y una matriz de covarianza representada por:

  [1.0 0.8
  0.8 1.0]

  Se pueden crear números aleatorios con una distribución normal multidimensional
  usando .np.random.multivariate_normal()
  La forma del ndarray creado será (500, 2).
  Para que el valor no cambie entre ejecuciones, np.random.seed ()
  Arreglemos el valor inicial aleatorio. Esta vez, usaré 0 como valor inicial.
'''

import numpy as np
import matplotlib.pyplot as plt

def generate_multivariate_data(seed=0, size=500):
    """
    Generate 2D multivariate normal data with fixed seed.

    Parameters:
        seed (int): Random seed value
        size (int): Number of samples to generate

    Returns:
        ndarray: Generated data with shape (size, 2)
    """
    np.random.seed(seed)
    mean = [-3, 0]
    cov = [[1.0, 0.8],
           [0.8, 1.0]]
    data = np.random.multivariate_normal(mean, cov, size)
    return data

def compute_statistics(data):
    """
    Compute and print the mean and covariance matrix of the data.

    Parameters:
        data (ndarray): Input data of shape (n, 2)
    """
    mean = np.mean(data, axis=0)
    cov = np.cov(data.T)
    print(f"Mean vector:\n{mean}")
    print(f"Covariance matrix:\n{cov}")

# Generate and use the functions
data = generate_multivariate_data()

compute_statistics(data)

#print(data)
```

```python
'''
PROBLEMA 2 - VISUALIZACION MEDIANTE DIAGRAMA DE DISPERSION
Visualice los puntos de datos creados en el problema 1 con un diagrama de dispersión.
Diagrama de dispersión de matplotlib.  Puedes dibujar usando plt.scatter ().
'''

import numpy as np
import matplotlib.pyplot as plt

def generate_multivariate_data(seed=0, size=500):
    """
    Generate 2D multivariate normal data with fixed seed.

    Parameters:
        seed (int): Random seed value
        size (int): Number of samples to generate

    Returns:
        ndarray: Generated data with shape (size, 2)
    """
    np.random.seed(seed)
    mean = [-3, 0]
    cov = [[1.0, 0.8],
           [0.8, 1.0]]
    data = np.random.multivariate_normal(mean, cov, size)
    return data

def visualize_data(data):
    """
    Visualize 2D multivariate data using a scatter plot.

    Parameters:
        data (ndarray): Input data of shape (n, 2)
    """
    plt.figure(figsize=(7, 6))
    plt.scatter(data[:, 0], data[:, 1], alpha=0.6, color='teal', edgecolor='k')
    plt.title("2D Multivariate Normal Distribution")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.grid(True, linestyle='--', linewidth=0.5)
    plt.axis("equal")
    plt.tight_layout()
    plt.show()

def compute_statistics(data):
    """
    Compute and print the mean and covariance matrix of the data.

    Parameters:
        data (ndarray): Input data of shape (n, 2)
    """
    mean = np.mean(data, axis=0)
    cov = np.cov(data.T)
    print(f"Mean vector:\n{mean}")
    print(f"Covariance matrix:\n{cov}")

# Generate and use the functions
data = generate_multivariate_data()
visualize_data(data)
compute_statistics(data)
```

```python
'''
PROBLEMA 3 - VISUALIZACION POR HISTOGRAMA
Se puede dibujar un histograma con plt.hist().
Cree uno para cada dimensión y conviértalo en dos histogramas.
El rango del eje horizontal de cada histograma debe alinearse con plt.xlim().
'''

import numpy as np
import matplotlib.pyplot as plt

def plot_histograms(data, bins=30):
    """
    Plot histograms for each dimension of 2D data.

    Parameters:
        data (ndarray): 2D array of shape (n, 2)
        bins (int): Number of histogram bins
    """
    x_data = data[:, 0]
    y_data = data[:, 1]

    # Determine global x-axis limits for consistency
    min_val = min(np.min(x_data), np.min(y_data))
    max_val = max(np.max(x_data), np.max(y_data))

    # Create figure with two subplots
    plt.figure(figsize=(10, 5))

    # Histogram for first dimension
    plt.subplot(1, 2, 1)
    plt.hist(x_data, bins=bins, color='skyblue', edgecolor='black')
    plt.title("Histogram of X Dimension")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.xlim(min_val, max_val)
    plt.grid(True, linestyle='--', linewidth=0.5)

    # Histogram for second dimension
    plt.subplot(1, 2, 2)
    plt.hist(y_data, bins=bins, color='salmon', edgecolor='black')
    plt.title("Histogram of Y Dimension")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.xlim(min_val, max_val)
    plt.grid(True, linestyle='--', linewidth=0.5)

    plt.tight_layout()
    plt.show()

# Example use with previously generated data
data = generate_multivariate_data()
plot_histograms(data)
```

```python
'''
PROBLEMA 4 - ADICION DE DATOS
Cree 500 nuevos números aleatorios con una distribución normal bidimensional,
con una media de (0, -3) y una matriz de covarianza representada por:

[1.0 0.8
 0.8 1.0]

Luego, visualice los datos de cada uno de los Problemas 1 y 4 como un único diagrama de dispersión.
Como leyenda, muestre el CERO en el Problema 1 y el UNO en el Problema 4.
'''

import numpy as np
import matplotlib.pyplot as plt

def generate_multivariate_data(seed, mean):
    """
    Generate 2D multivariate normal data with fixed seed and mean.

    Parameters:
        seed (int): Random seed value
        mean (list): Mean vector [x, y]

    Returns:
        ndarray: Generated data with shape (500, 2)
    """
    np.random.seed(seed)
    cov = [[1.0, 0.8],
           [0.8, 1.0]]
    data = np.random.multivariate_normal(mean, cov, 500)
    return data

def plot_two_datasets(data0, data1):
    """
    Plot two 2D datasets on the same scatter plot with legend.

    Parameters:
        data0 (ndarray): First dataset
        data1 (ndarray): Second dataset
    """
    plt.figure(figsize=(8, 6))
    plt.scatter(data0[:, 0], data0[:, 1], label="CERO", color='teal', alpha=0.6, edgecolor='k')
    plt.scatter(data1[:, 0], data1[:, 1], label="UNO", color='orange', alpha=0.6, edgecolor='k')
    plt.title("Multivariate Normal Distributions: CERO vs UNO")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.grid(True, linestyle='--', linewidth=0.5)
    plt.legend()
    plt.axis("equal")
    plt.tight_layout()
    plt.show()

# Generate datasets
data_cero = generate_multivariate_data(seed=0, mean=[-3, 0])
data_uno = generate_multivariate_data(seed=1, mean=[0, -3])

# Plot both datasets together
plot_two_datasets(data_cero, data_uno)
```

```python
'''
PROBLEMA 5 - COMBINACION DE DATOS
Es más fácil gestionar los datos posteriormente si se recopilan.
Combine los ndarrays creados en los problemas 1 y 4 para formar el ndarray de (1000, 2).
La unión se puede realizar usando ya.np.concatenate()np.vstack()
'''

import numpy as np

def generate_multivariate_data(seed, mean):
    """
    Generate 2D multivariate normal data with fixed seed and mean.

    Parameters:
        seed (int): Random seed value
        mean (list): Mean vector [x, y]

    Returns:
        ndarray: Generated data with shape (500, 2)
    """
    np.random.seed(seed)
    cov = [[1.0, 0.8],
           [0.8, 1.0]]
    data = np.random.multivariate_normal(mean, cov, 500)
    return data

def combine_data_concatenate(data1, data2):
    """
    Combine two datasets using np.concatenate() along axis 0.

    Parameters:
        data1 (ndarray): First dataset of shape (n, 2)
        data2 (ndarray): Second dataset of shape (n, 2)

    Returns:
        ndarray: Combined dataset of shape (2n, 2)
    """
    return np.concatenate((data1, data2), axis=0)

def combine_data_vstack(data1, data2):
    """
    Combine two datasets using np.vstack().

    Parameters:
        data1 (ndarray): First dataset of shape (n, 2)
        data2 (ndarray): Second dataset of shape (n, 2)

    Returns:
        ndarray: Combined dataset of shape (2n, 2)
    """
    return np.vstack((data1, data2))

# Generate original datasets
data_cero = generate_multivariate_data(seed=0, mean=[-3, 0])
data_uno = generate_multivariate_data(seed=1, mean=[0, -3])

# Combine using both methods
combined_data_concat = combine_data_concatenate(data_cero, data_uno)
combined_data_vstack = combine_data_vstack(data_cero, data_uno)

# Print shapes to confirm
print(f"Shape using concatenate: {combined_data_concat.shape}")
print(f"Shape using vstack:      {combined_data_vstack.shape}")
```

```python
'''
PROBLEMA 6 - ETIQUETADO
Para cada uno de los 1000 datos recopilados,
etiquételo para indicar si se creó en la pregunta 1 o 4.
El del problema 1 corresponde ZERO al del problema 4 corresponde ONE.
Agregue una nueva columna con etiquetas que contengan 0 y 1 y cree una matriz ndar de (1000, 3).
'''

import numpy as np

def label_and_combine(data0, data1):
    """
    Label and combine two datasets into a single ndarray with shape (1000, 3).

    Parameters:
        data0 (ndarray): Dataset from Problem 1 (label 0), shape (500, 2)
        data1 (ndarray): Dataset from Problem 4 (label 1), shape (500, 2)

    Returns:
        ndarray: Combined and labeled dataset, shape (1000, 3)
    """
    # Create label arrays
    labels0 = np.zeros((data0.shape[0], 1), dtype=int)
    labels1 = np.ones((data1.shape[0], 1), dtype=int)

    # Add labels as third column
    labeled_data0 = np.hstack((data0, labels0))
    labeled_data1 = np.hstack((data1, labels1))

    # Combine both datasets
    combined = np.vstack((labeled_data0, labeled_data1))
    return combined

# Generate original datasets again (if needed)
data_cero = generate_multivariate_data(seed=0, mean=[-3, 0])
data_uno = generate_multivariate_data(seed=1, mean=[0, -3])

# Combine with labels
labeled_combined_data = label_and_combine(data_cero, data_uno)

# Print shape and first 5 rows to confirm
print(f"Combined shape: {labeled_combined_data.shape}")
print("Labeled data:")
#print(labeled_combined_data[:10])
print(labeled_combined_data)
```
