```python
'''
PROBLEMA 1: Cree un programa que use operadores aritméticos de exponenciación.

'''

thickness = 0.0008      #grosor de papel
n_folds = 43            #nro de veces a doblar

folded_thickness = thickness * 2 ** n_folds

print("espesor: {} metros".format(folded_thickness))
```

```python
'''
PROBLEMA 2: convertir de metros a km y visualizar el resultado
'''
thickness = 0.0008      #grosor de papel
n_folds = 43            #nro de veces a doblar

folded_thickness = thickness * 2 ** n_folds / 1000

print("espesor: {:.2f} kilometros".format(folded_thickness))
```

```python
'''
PROBLEMA 3: cree un programa que use la instrucción for con operacions aritmeticas basicas
'''
thickness = 0.0008      #grosor de papel sin doblar
n_folds = 43            #nro de veces a doblar

folded_thickness = thickness / 1000       #grosor del papel doblado

for i in range(n_folds):
    folded_thickness = folded_thickness * 2


print("espesor: {:.2f} kilometros".format(folded_thickness))
```

```python
'''
PROBLEMA 4: crear un programa donde se midan los tiempos de ambos metodos de calculo
'''
import time

thickness = 0.0008      #grosor de papel sin doblar
n_folds = 43            #nro de veces a doblar

#metodo directo
start1 = time.time()
folded_thickness1 = thickness * 2 ** n_folds / 1000
elapsed_time1 = time.time() - start1

print("espesor: {:.2f} kilometros".format(folded_thickness1))
print("elapsed time 1st metod: {} seconds".format(elapsed_time1))

#metodo de ciclos
start2 = time.time()
folded_thickness2 = thickness / 1000       #grosor del papel doblado
for i in range(n_folds):
    folded_thickness2 = folded_thickness2 * 2

elapsed_time2 = time.time() - start2

print("espesor: {:.2f} kilometros".format(folded_thickness2))
print("elapsed time 2nd metod: {} seconds".format(elapsed_time2))
```

```python
'''
PROBLEMA 5: crear un programa que almacene en una lista el resultado de cada ciclo
'''

thickness_paper = 0.00008                                             #grosor expresado en metros
n_folds = 43                                                           #nro de veces a doblar
folded_thickness = thickness_paper                                     #grosor actual del papel doblado
thickness_list = [thickness_paper]

for i in range(n_folds):
    folded_thickness = folded_thickness * 2
    thickness_list.append(folded_thickness)

print(f"La longitud de la lista es: {len(thickness_list)}")
```

```python
'''
PROBLEMA 6: crear un programa que visualice el grafico de los calculos
'''
import matplotlib.pyplot as plt

thickness_paper = 0.00008                                             #grosor expresado en metros
n_folds = 43                                                           #nro de veces a doblar
folded_thickness = thickness_paper                                     #grosor actual del papel doblado
thickness_list = []

for i in range(n_folds):
    folded_thickness = folded_thickness * 2
    thickness_list.append(folded_thickness)

#visualizacion de los resultados de calculo
print("espesor: {} metros".format(folded_thickness))
print("espesor: {:.2f} kilometros".format(folded_thickness / 1000))

# Visualizar gráfico.Con el título y los nombres de las etiquetas de los ejes.
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(thickness_list)
plt.show()
```
