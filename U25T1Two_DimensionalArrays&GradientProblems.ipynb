```python
'''
PROBLEMA 1 - FUNCION LINEAL
  Considere la siguiente fórmula donde X está en el rango de -50 a 50.
  y=12x+1
  Cree un ndarray que represente (X) en intervalos de -50 a 50
  y un ndarray que represente (Y) para cada elemento.
'''

import numpy as np

# Crear el ndarray de X en el rango de -50 a 50 (inclusive)
X = np.arange(-50, 51)

# Calcular Y = 12X + 1 para cada valor en X
Y = 12 * X + 1

# Ver los resultados
print("X array:", X)
print("Y array:", Y)
```

```python
'''
PROBLEMA 3 - ENCONTRAR LA GRADIENTE
  Encuentra la pendiente en cada coordenada x.
  Aquí la pendiente se calcula dividiendo el cambio en "y" por el cambio en "x".
    Gradient = (y2 - y1) / (x2 - x1)
  nota:
    Es necesario usar el límite (derivado) para obtener el gradiente en cada punto exactamente
    Gradient usa el espaciado de coordenadas x, por lo que uno menos para x elementos
'''
import numpy as np

# Problema 1
x = np.arange(-50, 51)       # x va de -50 a 50, total 101 elementos
y = 12 * x + 1               # función lineal

# Problema 3
dy = y[1:] - y[:-1]          # diferencia de y
dx = x[1:] - x[:-1]          # diferencia de x
gradient = dy / dx           # pendiente en cada intervalo

print("Gradient array:" ,gradient)
```

```python
'''
PROBLEMA 4 - DIBUJA UN GRAFICO
  Dibuja una gráfica de la función lineal usando Matplotlib.
  Además, dibuje un gráfico con x en el eje horizontal y gradiente en el eje vertical.
'''
import numpy as np
import matplotlib.pyplot as plt

# Datos base
x = np.arange(-50, 51)
y = 12 * x + 1

# Gradiente
dy = y[1:] - y[:-1]
dx = x[1:] - x[:-1]
gradient = dy / dx

# Para igualar longitud con x, agregamos un valor extra al gradiente
# Aquí repetimos el último valor para tener 101 elementos
gradient_full = np.append(gradient, gradient[-1])

# Dibujo de gráfico
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Función y = 12x + 1', color='blue')
plt.plot(x, gradient_full, label='Gradiente', color='orange', linestyle='--')

# Detalles del gráfico
plt.title('Función Lineal y su Gradiente')
plt.xlabel('x')
plt.ylabel('Valor')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
'''
PROBLEMA 5 - FUNCIONALIZACION DE PYTHON
  Consulte la plantilla para crear una función de Python que resuma el rango del problema 1 al problema 3.
  Luego, calcule el gradiente de las siguientes tres ecuaciones a intervalos de 0.1 y póngalo como ndarray.

  y= x^2            (−50 ≤ x ≤ 50)
  y= 2x^2 + 2^x     (−50 ≤ x ≤ 50)
  y= sin(x^(1/2)    (0 ≤ x ≤ 50)

  Además, dibuje el gráfico como en la pregunta 4.
  Para calcular la función trigonométrica sin para ndarray. Es usado np.sin ()
'''
import numpy as np
import matplotlib.pyplot as plt

def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    """
    Computes the gradient using differences.
      Parameters
      ----------------
      function : function
        The function to evaluate. Takes an ndarray x and returns ndarray y.
      x_range : tuple
        Range for np.arange() to generate x values.

      Returns
      ----------------
      array_xy : ndarray, shape(n, 2)
        Combined x and y values. n depends on x_range.
      gradient : ndarray, shape(n-1,)
        The gradient of the function. One element less than x.
    """
    x = np.arange(*x_range)
    y = function(x)
    array_xy = np.stack((x, y), axis=1)
    gradient = np.diff(y) / np.diff(x)
    return array_xy, gradient

def function1(x):
    return x ** 2

def function2(x):
    return 2 * x ** 2 + 2 ** x

def function3(x):
    return np.sin(np.sqrt(x))

# Apply the compute_gradient function to each
array1, grad1 = compute_gradient(function1)
array2, grad2 = compute_gradient(function2)
array3, grad3 = compute_gradient(function3, x_range=(0, 50.1, 0.1))

# Plotting
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

# Plot 1
axs[0].plot(array1[:, 0], array1[:, 1], label="y = x^2")
axs[0].plot(array1[:-1, 0], grad1, label="Gradient", linestyle="--")
axs[0].set_title("Function 1: y = x^2")
axs[0].legend()
axs[0].grid()

# Plot 2
axs[1].plot(array2[:, 0], array2[:, 1], label="y = 2x^2 + 2^x")
axs[1].plot(array2[:-1, 0], grad2, label="Gradient", linestyle="--")
axs[1].set_title("Function 2: y = 2x^2 + 2^x")
axs[1].legend()
axs[1].grid()

# Plot 3
axs[2].plot(array3[:, 0], array3[:, 1], label="y = sin(sqrt(x))")
axs[2].plot(array3[:-1, 0], grad3, label="Gradient", linestyle="--")
axs[2].set_title("Function 3: y = sin(sqrt(x))")
axs[2].legend()
axs[2].grid()

plt.tight_layout()
plt.show()

```

```python
'''
PROBLEMA 6 - BUSQUEDA DEL VALOR MINIMO
  Encuentre valor mínimo de "Y" para las funciónes tratadas en el problema 5.
  El rango de "X" es el mismo que en la pregunta 5.
  Esta vez, utilice el método ndarray min y el método argmin en lugar del método de descenso de gradiente.
  Además, muestre la pendiente antes y después de tomar el mínimo "Y".
'''
import numpy as np
import matplotlib.pyplot as plt

# Define the compute_gradient function
def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    """
    Calculate the gradient using difference of values.

    Parameters
    ----------------
    function : function
        The function to evaluate, which returns y as ndarray when x is ndarray.
    x_range : tuple
        Range used in np.arange(start, stop, step)

    Returns
    ----------------
    array_xy : ndarray, shape(n, 2)
        Combined x and y values as columns.
    gradient : ndarray, shape(n-1,)
        Gradient values. One less than the number of x values.
    """
    x = np.arange(*x_range)
    y = function(x)
    array_xy = np.stack([x, y], axis=1)
    gradient = np.diff(y) / np.diff(x)
    return array_xy, gradient

# Define the three functions
def function1(x):
    return x ** 2

def function2(x):
    return 2 * x ** 2 + 2 ** x

def function3(x):
    return np.sin(np.sqrt(x))

# Calculate for each function
functions = [function1, function2, function3]
names = ['y = x^2', 'y = 2x^2 + 2^x', 'y = sin(sqrt(x))']
x_ranges = [(-50, 50.1, 0.1), (-50, 50.1, 0.1), (0, 50.1, 0.1)]

for i, func in enumerate(functions):
    # Get x and y values, and gradient
    array_xy, gradient = compute_gradient(func, x_ranges[i])
    x = array_xy[:, 0]
    y = array_xy[:, 1]

    # Find minimum y value and its index
    min_index = y.argmin()
    min_x = x[min_index]
    min_y = y[min_index]

    # Show gradient before and after the minimum
    if 0 < min_index < len(gradient) - 1:
        grad_before = gradient[min_index - 1]
        grad_after = gradient[min_index]
    else:
        grad_before = grad_after = None

    # Print results
    print(f'Function: {names[i]}')
    print(f'Minimum y value: {min_y:.4f} at x = {min_x}')
    print(f'Gradient before minimum: {grad_before}')
    print(f'Gradient after minimum: {grad_after}\n')

    # Plot the function and gradient
    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label='Function')
    plt.plot(x[:-1], gradient, label='Gradient', linestyle='--')
    plt.axvline(min_x, color='red', linestyle=':', label='x of min y')
    plt.title(f'Function and Gradient: {names[i]}')
    plt.xlabel('x')
    plt.ylabel('y / gradient')
    plt.legend()
    plt.grid(True)
    plt.show()

```
